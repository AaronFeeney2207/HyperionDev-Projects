Use the data to determine how age affects insurance costs:

- Plot a scatterplot with age on the x-axis and charges on the y-axis 
- Useing linear_model.LinearRegression() from sklearn, fit a model to your data, and make predictions on data 
- Plot another scatter plot with the best fit line

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
column_data = pd.read_csv('insurance.csv')
column_data.head()
```

```python
# Plotting a scatterplot with relevant colums of data 

x = column_data.loc[:,'age'].values
y = column_data.loc[:,'charges'].values


plt.scatter(x,y,color='r')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

```python
# Shaping model 

x = column_data.loc[:,'age'].values.reshape(-1, 1)
y = column_data.loc[:,'charges'].values

# Fitting the model 

insurance_model = LinearRegression()
insurance_model.fit(x,y)


# Data and model 
y_pred = insurance_model.predict(x)
plt.scatter(x,y,color='r')
plt.plot(x,y_pred,color='g')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

Can use this model to predict what the approximate charges would be on an age that is outside of our data range

```python
# Get the max age from our dataset

column_data.loc[:,'age'].max()
```

```python
# Predicting unknown ages that wasn't in our original dataset 

unknown_age = [[75]]

x_pred = np.append(x,unknown_age).reshape(-1, 1)
y_pred = insurance_model.predict(x_pred)

plt.scatter(x,y,color='r')
plt.plot(x_pred,y_pred,color='g')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()

print(f'Charges for a 75 year old should approximately be: {insurance_model.predict(unknown_age)}')
```

We can conclude that there is some strength to the statement that the higher the age, the higher the charges will be based on these two variables alone and our model that portrays the best fit line.
